```python
def generate_rapidities(Nch, sigma):
    #Generate the rapidity distribution of charged particles
    
    #The charged particle rapidities are distributed according to a normal distribution for a single event.
    
    # Nch is the number of charged particles
    #sigma is the width of the normal distribution which may give infor about longitudinal expansion
        
    #y is the rapidity of each charged particle
    
    from numpy.random import normal
    
    y = normal(0,sigma,size=Nch)
    
    return y
```

```python
def dytodeta(y,a=.938/2):
    # Want to transforms rapidity into pseudorapidity 
     
 
    # Ref ch. 8 Statistics Overview by Chrisitan Holm Christensen 
    #Make an array of the rapidity y
    
    # a is the m/pT ratio of particles       
  
    # get back eta which is an array of the pseudorapidity eta
    from numpy import tanh , sqrt , log, where
    
    t = tanh(y)
    tmp = -((1.+2.*a**2)*t**2 - 2.*t*sqrt(a**2+1.)*sqrt(a**2*t**2+1.) + 1.)/(t**2-1.)
    #eta = -0.5 * log(where(tmp > 0, tmp, 1e-11)) 
    eta = -0.5 * log(where(tmp > 0, tmp, 1e-8))  
    #eta = -0.5 * log(tmp[tmp >0]) 
        
    return eta 
```

```python
def ytoeta(Nch=15000):
    import numpy
    from numpy import linspace
    import matplotlib
    import matplotlib.pyplot as plt

    #creates bin range and number of bins
    bins = linspace(-5,5,26)

    #get rapidities Nch=15000
    y = generate_rapidities(Nch=Nch,sigma=4)
    
    #converting y to eta
    eta = dytodeta(y,a=0.938/2) #used the mass of a proton, will probably used higher momentum later on
    
    from nbi_stat import histogram, plot_hist
    
    #make histogram of eta distribution
    heta = histogram(eta,bins=bins ,normalize=False)
    arr = numpy.asarray(heta)
    numpy.savetxt("Nch15.csv", arr, delimiter=",") #save to CSV for sample data array --> want to use in root
    
    from matplotlib.pyplot import figure, gca, plot, xlabel, ylabel, legend, title
    
    #plotting
    fig = figure(dpi=100)
    ax = fig.add_subplot() #used to add text onto the plots
    plot_hist(*heta ,fmt='s',label=r'$\mathrm{N_{ch}=15000}$',color='C2');
    ax.text(0.99, 0.87, '$\mathrm{p_{T}=2 GeV}$',
        verticalalignment='top', horizontalalignment='right',
        transform=ax.transAxes,
        color='black', fontsize=10)
    ax.text(0.99, 0.82, '$\mathrm{m_{p}=0.938 MeV}$',
        verticalalignment='top', horizontalalignment='right',
        transform=ax.transAxes,
        color='black', fontsize=10)
    xlabel(r"$\eta$")
    ylabel(r"$\frac{\mathrm{d}N_{\mathrm{ch}}}{\mathrm{d}\eta}$",loc='top',rotation=0)
    legend();
    title("Central Plateau in 0+1D Hydrodynamics")
    
    
    return
ytoeta();
```
